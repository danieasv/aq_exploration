```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import math
import numpy as np
import plotly.express as px
#warnings.filterwarnings('ignore')
```

```python
full_table = pd.read_csv('./aq_dump.csv',sep=',',parse_dates=['received_time'])
full_table = full_table.drop_duplicates(subset="received_time", keep=False)
full_table['received_time'] = pd.to_datetime(full_table['received_time'], unit='ms')
full_table = full_table.rename(columns={"board_rel_hum": "Relative Humidity", "received_time" : "Time"})

# Convert from radians to degrees
rad_to_deg = 180/math.pi
full_table['lat'] = rad_to_deg * full_table['lat']
full_table['lon'] = rad_to_deg * full_table['lon']


# Add a column with device names
sverresborg =  full_table["device_id"] == "17dh0cf43jg77j"
berg =  full_table["device_id"] == "17dh0cf43jg783"
singsaker =  full_table["device_id"] == "17dh0cf43jg781"
bispehaugen = full_table["device_id"] == "17dh0cf43jg77n"
telenor = full_table["device_id"] =='17dh0cf43jg6n4'
full_table['Names'] = np.select([sverresborg,berg,singsaker, bispehaugen, telenor], ['Sverresborg skole','Berg skole','Singsaker skole', 'Bispehaugen skole', 'Telenor'], default=-1)

full_table.head()
```
